```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
import zipfile

# Load the dataset (update the path if needed)
file_path = "C:\\Users\\Renuka\\Desktop\\Skillcraft\\Dataset\\skillcraftdata.zip"
csv_name = "API_SP.POP.TOTL_DS2_en_csv_v2_763389.csv"
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path}\nPlease check the path and ensure the file exists.")

with zipfile.ZipFile(file_path) as z:
    with z.open(csv_name) as f:
        df = pd.read_csv(f, skiprows=4)

# Select population data for the latest year (2024)
pop_2024 = df[['Country Name', '2024']].dropna()
pop_2024.columns = ['Country', 'Population']

# -------------------------------
# Bar Chart: Top 10 most populated countries in 2024
# -------------------------------
top10 = pop_2024.sort_values(by='Population', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10['Country'], top10['Population'] / 1e9, color='skyblue', edgecolor='black')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Population (Billions)')
plt.title('Top 10 Most Populated Countries in 2024')
plt.show()

# -------------------------------
# Histogram: Distribution of populations in 2024
# -------------------------------
plt.figure(figsize=(10,6))
plt.hist(pop_2024['Population'] / 1e6, bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Population (Millions)')
plt.ylabel('Number of Countries')
plt.title('Distribution of Country Populations in 2024')
plt.show()
```
